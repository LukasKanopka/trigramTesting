```python
import torch
```

```python
words = open('names.txt').read().splitlines()

#print(words[:10])

b = {}
for w in words:
  chs = ['.'] + list(w) + ['.']
  for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
      #print(ch1, ch2, ch3)
      trigram = (ch1, ch2, ch3)
      b[trigram] = b.get(trigram, 0) + 1

print(sorted(b.items(), key = lambda kv: -kv[1]))

N = torch.zeros((27, 27, 27), dtype=torch.int32)

chars = sorted(list(set(''.join(words))))
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s,i in stoi.items()}

for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2, ch3 in zip(chs, chs[1:], chs[2:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        ix3 = stoi[ch3]
        N[ix1, ix2, ix3] += 1
```

```python
N2 = torch.zeros((27, 27), dtype=torch.int32)
for x in range(27):
    for y in range(27):
        sum = 0
        for element in N[x, y]:
            sum += element
        N2[x, y] = sum

N2[0]
```

```python
g = torch.Generator().manual_seed(2147483148)
for i in range(20):
    out = ""
    ix = 0
    ix2 = torch.multinomial(p1, num_samples=1, replacement=True, generator=g).item()
    out += (itos[ix2])
    while True:
        p = N[ix, ix2].float()
        p = p / p.sum()
        prediction = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        out += (itos[prediction])
        if prediction == 0:
            break
        ix = ix2
        ix2 = prediction

    print(out)
```
